# 数据加载
在pytorch中, 数据加载可通过自定义的数据集对象. 数据集对象被抽象为Dataset类, 实现自定义的数据集需要集成Dataset, 并实现两个python魔法方法:
* __getitem__: 返回一条数据, 或一个样本. obj[index]等价于obj.__getitem__(index)
* __len__: 返回样本的数量. len(obj)等价于obj.__len__()

下面以dogs vs cats为举例

```python
%env LS_COLORS = None
!tree --charset ascii data/dogcat/
```

```python
import torch as t
from torch.utils import data
```

```python
import os 
from PIL import Image
import numpy as np

class DogCat(data.Dataset):
    def __init__(self, root):
        '''
        这里不加载图片, 只是指定图片路径, 当调用__getitem__时才会真正读图片
        '''
        imgs = os.listdir(root)
        self.imgs = [os.path.join(root, img) for img in imgs]
    def __getitem__(self, index):
        img_path = self.imgs[index]
        label = 1 if 'dog' in img_path.split('/')[-1] else 0
        pil_img = Image.open(img_path)
        array = np.asarray(pil_img)     # numpy
        data = t.from_numpy(array)      # 转化为tensor
        return data, label
    def __len__(self):
        return len(self.imgs)
```

```python
dataset = DogCat('./data/dogcat')
img, label = dataset[0]     # 相当于调用dataset.__getitem__(0)
for img, label in dataset:
    print(img.size(), label)
```

通过上面的代码, 我们学习了如何自定义自己的数据集, 并可以依次获取. 但这里返回的数据不适合实际使用, 因为:
* 返回形状不一样, 对于需要取batch的网络很不友好
* 返回样本的数值较大, 未归一化至[-1, 1]

针对上述问题, torch提供torchvision, 它提供了很多计算机视觉的图像处理工具.

对PIL Image的操作包括:
* Scale: 调整图片尺寸, 长宽比保持不变
* CenterCrop, RandomCrop, RandomResizedCrop: 裁剪图片
* Pad
* ToTensor: 将PIL Image转化为Tensor, 会自动将[0, 255]归一化至[0, 1], 先将PIL Image转为Tensor, 再将(H, W, C)转化为(C, H, W), 最后再除以255.

对Tensor的操作包括:
* Normalize: 标准化, 减均值, 除以标准差
* ToPILImage: 将Tensor转为PIL Image

如果要对图片进行多个操作，可通过Compose函数将这些操作拼接起来, 类似于nn.Sequential. 注意, 这些操作定义后是以函数的形式存在, 真正使用时需要调用它的__call__方法, 这点有点类似于nn.Module.

```python
import os 
from PIL import Image
import numpy as np
from torchvision import transforms as T

transform = T.Compose([
    T.Resize(224),               # 缩放图片, 保持长宽比不变, 最短边为224
    T.CenterCrop(224),           # 将图片中间切除224*224的图片
    T.ToTensor(),                # 将图片转为Tensor, 归一化至[0, 1]
    T.Normalize(mean=[.5, .5, .5], std=[.5, .5, .5])     # 标准化至[-1, 1]
])
class DogCat(data.Dataset):
    def __init__(self, root, transforms=None):
        '''
        这里不加载图片, 只是指定图片路径, 当调用__getitem__时才会真正读图片
        '''
        imgs = os.listdir(root)
        self.imgs = [os.path.join(root, img) for img in imgs]
        self.transforms = transforms
    def __getitem__(self, index):
        img_path = self.imgs[index]
        label = 1 if 'dog' in img_path.split('/')[-1] else 0
        pil_img = Image.open(img_path)
        if self.transforms:
            data = self.transforms(pil_img)
        return data, label
    def __len__(self):
        return len(self.imgs)
```

```python
dataset = DogCat('./data/dogcat', transform)
img, label = dataset[0]     # 相当于调用dataset.__getitem__(0)
for img, label in dataset:
    print(img.size(), label)
```

torchvision中常用到的Dataset是ImageFolder. ImageFolder假设所有的文件按文件夹保存, 每个文件夹存储同一个类别的图片, 文件夹名为类名, 其构造函数如下:

ImageFolder(root, transform=None, target_transform=None, loader=default_loader)

它只要有4个参数:
* root: 在root指定的路径下寻找图片
* transform: 对PIL Imag进行的转换操作, transform的输入是使用loader读取图片的返回对象
* target_transform: 对label进行转换
* loader: 给定路径后如何读取图片, 默认读取为RGB格式的PIL Image对象

label是按照文件夹顺序排序后存成的字典, 即{类名: 类序号(从0开始)}, 一般来说最好直接将文件夹命名为从0开始的数字, 这样会和ImageFolder实际的label一致, 如果不是这种命名规范, 建议看看self.class_to_idx属性以了解label的文件夹名的映射关系

Dataset只负责数据的抽象, 一次调用__getitem__只返回一个样本. 在训练神经网络时需要一个batch, 同时需要对数据进行shuffle和并行加速等. pytorch提供了DataLoader

DataLoader的函数定义如下: DataLoader(dataset, batch_size=1, shuffle=False, sampler=None, num_workers=0, collate_fn=default_collate, pin_memory=False, drop_last=False)
* dataset: 加载的数据集(Dataset对象)
* batch_size
* shuffle: 是否打乱
* sampler: 样本抽样
* num_workers: 使用多进程加载的进程数
* collate_fn: 如何将多个样本数据拼接成一个batch, 一般使用默认的拼接方式即可
* pin_memory: 是否将数据保存在pin memory区, pin_memory中的数据转到GPU会快一些
* drop_last: dataset中的数据个数可能不是batch_size的整数倍, drop_last为True会将多出来不足一个batch的数据丢弃

在数据处理中, 有时会出现某个样本无法读取等问题, 比如图片损坏. 这时在__getitem__函数中将出现异常, 此时最好的解决方案是将出错的样本删除/如果实在遇到这种清空无法处理, 可返回None对象, 然后在DataLoader中实现自定义的collate_fn, 将空对象过滤掉. 这时dataloader返回的样本数小于batch_size

```python
class NewDogCat(DogCat):
    def __getitem__(self, index):
        try:
            # 调用父类的获取函数, 即DogCat.__getitem__(self, index)
            return super(NewDogCat, self).__getitem__(index)
        except:
            return None, None

from torch.utils.data.dataloader import default_collate
def my_collate_fn(batch):
    batch = list(filter(lambda x: x[0] is not None, batch))
    if len(batch) == 0:
        return t.Tensor()
    return default_collate(batch)     # 用默认方式拼接过滤后的batch数据
```

```python
dataset = NewDogCat('data/dogcat_wrong/', transforms=transform)
```

```python
dataset[5]
```

```python
dataloader = DataLoader(dataset, 2, collate_fn=my_collate_fn, num_workers=1, shuffle=True)
for batch_datas, batch_labels in dataloader:
    print(batch_datas.size(), batch_labels.size())
```

对于样本损坏或数据集加载异常等情况, 还可以通过其他方式解决. 例如凡是遇到异常清空, 就随机取一张图片代替

```python
class NewDogCat(DogCat):
    def __getitem__(self, index):
        try:
            # 调用父类的获取函数, 即DogCat.__getitem__(self, index)
            return super(NewDogCat, self).__getitem__(index)
        except:
            new_index = random.randint(0, len(self-1))
            return self[new_index]
```

对于丢弃异常图片而言, 这种做法更好一点, 因为保证每个batch的数目还是batch_size个. 但在大多数情况下, 最好的方式还是对数据进行清洗.

对DataLoader和Dataset的使用建议:
1. 将高负载的操作放在__getitem__中, 如加载图片等
2. dataset中应尽量包含只读对象, 避免修改任何可变对象, 利用多线程加速

# Visdom
visdom中有两个重要的概念:
* env: 环境. 不同环境的可视化结果相互隔离, 互不影响, 在使用时如果不指定env, 默认使用main. 不同用户, 不同程序一般使用不同的env.
* pane: 窗格. 窗格可用于可视化图像, 数值或打印文本等, 可以缩放拖动保存和关闭. 一个程序中可使用同一个env中的不同pane, 每个pane可视化或记录某一信息.

Visdom需要通过 python -m visdom.server命名启动visdom服务, 或通过nohup python -m visdom.serve &命名将服务放至后台.

Visdom的使用需要注意两点:
* 需手动指定保存env, 可在web界面点解save按钮或在程序中调用save方法, 否则visdom服务重启后, env等信息会丢失.
* 客户端与服务器之间的交互采用tornado异步框架, 可视化操作不会阻塞当前程序, 网络异常也不会导致程序退出.

```python
import torch as t
```

```python
import visdom

# 新建一个连接客户端
# 制定env=u'test1', 默认端口为8097, host是'localhost'
vis = visdom.Visdom(env=u'test1', use_incoming_socket=False)

x = t.arange(1, 30, 0.01)
y = t.sin(x)
vis.line(X=x, Y=y, win='sinx', opts={'title':'y=sin(x)'})
```

vis作为一个客户端对象, 可以使用常见的画图函数, 包括:
* line: 类似于plot, 用于记录标量变化, 如损失, 准确率等
* image: 可视化图片, 可以是输入图片, 也可以是GAN生成图片, 还可以是卷积核信息
* text: 用于记录日志等文字信息
* histgram: 可视化分布, 主要是查看数据,参数的分布
* scatter: 绘制散点图
* bar: 绘制柱状图

Visdom同时支持pytorch的tensor和numpy的ndarray两种数据结构, 但不支持的Python的int, float等类型. 上述操作的参数一般不同, 但有两个参数是绝大多数操作都具备的:
* win: 用于指定pane的名字, 如果不指定, visdom将自动分配一个新的pane. 如果两次操作制定win的名字一样, 新的操作将覆盖当前pane的内容.
* opts: 选项, 接受一个字典, 常见的option包括title, xlabel, ylabel, width等, 主要用于设置pan的显示格式.

之前提到过, 每次操作都会覆盖之前的数值, 但往往我们在训练网络过程中需要不断更新数值, 如损失等, 这是就需要制定参数update='append'来避免覆盖之前的数值. 而除了使用updete参数以外, 还可以使用vis.updateTrace方法来更新图, 但updateTrace不仅能在制定pane上新增一个和已有数据互相独立的Trace, 还能想update='append'那样在同一条trace上追加数据.

```python
# append追加数据
for ii in range(10):
    x = t.Tensor([ii])
    y = x
    vis.line(X=x, Y=y, win='poly', update='append' if ii > 0 else None)
    
# updateTrace新增一条线
x = t.arange(0, 9, 0.1)
y = (x**2) / 9
vis.line(X=x, Y=y, win='poly',name='this is a new line', update='new')
```
